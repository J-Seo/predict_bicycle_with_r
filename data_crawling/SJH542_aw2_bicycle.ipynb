```python
import pandas as pd
```

```python
data = pd.read_csv('SJH542_(bicycle)d.csv')
data
```

```python
d1 = pd.read_excel('SJH542_(bicycle)org.xlsx')
d1
```

```python
d1['대여시간']
```

```python
i = 0
tmp = 0
start_point = 0
cnt_container = list()


for d in d1['대여시간']:
    
    
    if tmp != d:
        
        cnt = i - start_point + 1
        start_point = i + 1
        cnt_container.append(cnt)
        
    i += 1
    tmp = d
    
    if i == len(d1['대여시간']):
        cnt = i - start_point + 1
        cnt_container.append(cnt)
```

```python
len(cnt_container)
```

```python
data['cnt'] = cnt_container
```

```python
data
```

```python
data.to_csv('SJH542_(bicycle)d.csv', index = False, encoding = 'utf-8', header = True, sep = ',')
```
